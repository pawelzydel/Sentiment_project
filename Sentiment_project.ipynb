```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
df2 = pd.read_csv('/content/sentiment_ready.csv')
```

```python
grouped_data = df2.groupby(['Party', 'Sentiment']).size().unstack()

# Tworzenie wykresu
fig, ax = plt.subplots(figsize=(8, 6))
grouped_data.plot(kind='bar', ax=ax, width=0.8, edgecolor='black', alpha=0.8)

# Ustawienie etykiet i tytułu
ax.set_xticklabels(['Demokraci', 'Republikanie'], rotation=0)
ax.set_xlabel('Partia')
ax.set_ylabel('Liczba tweetów')
ax.set_title('Porównanie sentymentu między Demokratami a Republikanami')

# Legenda
ax.legend(["Negatywny", "Pozytywny"], loc="upper right")

# Wyświetlenie wykresu
plt.tight_layout()
plt.show()
```

```python
# Grupowanie danych na podstawie kolumn "Party" i "Sentiment"
grouped_data = df2.groupby(['Party', 'Sentiment']).size().unstack()

# Obliczenie sumy dla każdej kategorii
sum_by_party = grouped_data.sum(axis=1)

# Obliczenie procentowego udziału sentymentu w każdej kategorii
percentages = (grouped_data.T / sum_by_party).T * 100

# Sprawdzenie unikalnych wartości w kolumnie "Party"
unique_parties = df2['Party'].unique()

# Tworzenie wykresów kołowych
fig, axes = plt.subplots(1, len(unique_parties), figsize=(12, 6))

# Iteracja po unikalnych wartościach "Party" i generowanie wykresów
colors = ['#FF4F4F', '#66CC66']  # Kolor czerwony dla negatywnego, zielony dla pozytywnego sentymentu

for i, party in enumerate(unique_parties):
    ax = axes[i]
    ax.pie(percentages.loc[party], labels=['Negatywny', 'Pozytywny'], colors=colors, autopct='%1.1f%%', startangle=90)
    ax.set_title(f'Sentyment wśród {party}')

# Wyświetlenie wykresów kołowych
plt.tight_layout()
plt.show()
```
